```python
import pandas
from scipy.stats import binom
import matplotlib.pyplot as plt
import fractions


# Read the Excel file
df = pandas.read_excel('Computer.xlsx')

# Initialize the count variable
p = 0
total =0
# Iterate over the rows of the dataframe
for i in range(df.shape[0]):
    # Get the value of the 'Gender' column for the current row
    gender = df['Gender'][i]

    # Check if the value is 'F'
    if gender.strip().upper() == 'F':
        # Increment the count if the value is 'F'
        p += 1
    total+=1

print("The Pobabilty of Successes p : ",p/total)
print("The Pobabilty of Failure q : ",1-p/total)

print("P(X=20) ")

n = total
p = p/total
r =20

dist = binom.pmf(r, n, p)#Applying binomial distribution on data

# Convert a number to a fraction
fraction = fractions.Fraction.from_float(dist)
print("The Probability that 2023 class contain 20 females is : ",fraction)
```
