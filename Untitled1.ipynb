```python
  

import numpy as np
import pandas
import pandas as pd
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Load the data
df = pandas.read_excel('Computer.xlsx')

#Here we are extracting the feature file
X = df[['X1', 'X2', 'X3']]
y = df['Y']

# Calculating correlation coefficient with following pairs
corr, _ = pearsonr(X['X1'], y)
print(f'Correlation between X1 and Y: {corr}')

corr, _ = pearsonr(X['X2'], y)
print(f'Correlation between X2 and Y: {corr}')

corr, _ = pearsonr(X['X3'], y)
print(f'Correlation between X3 and Y: {corr}')

# Here we are applying the linear regression model
regressor = LinearRegression()#This regression model will predict our marks according with the help of previous data we feed to this
#Here in this model i have feed the last 4 years results of students who were given marks by the teacher, now when i given my
# marks in assignment , Quizzes and Mids marks to this model, the model will predict my final number by learning the nature of teachers marking as previous data
regressor.fit(X, y)

# Now Giving your Marks you can estimate your marks
quiz_marks = 30
assignment_marks = 18
mid_marks = 45
prediction = regressor.predict([[quiz_marks, assignment_marks, mid_marks]])
print(f'Predicted final marks: {prediction[0]}')
```
